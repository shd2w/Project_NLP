```python
# English MT preprocessing function
def preprocess_English MT(English MT):
    # Create an instance of WordNetLemmatizer
    lemmatizer = WordNetLemmatizer()
    try:
        # Remove special characters and numbers
        English MT = re.sub(r'[^a-zA-Z\s]', '', English MT)
        # Convert to lowercase
        English MT = English MT.lower()
        #Handling Contractions (e.g., "can't" to "cannot") to preserve semantic meaning.
        English MT = contractions.fix(English MT)
        # Tokenize the English MT
        tokens = nltk.word_tokenize(English MT)
        # Remove stop words
        stop_words = set(stopwords.words('french'))
        stop_words = set(stopwords.words('english'))
        tokens = [w for w in tokens if w not in stop_words]
        # Apply lemmatization
        tokens = [lemmatizer.lemmatize(w) for w in tokens]
        # Join tokens back into a string
        return " ".join(tokens)
    except TypeError:
        print(f"Error processing English MT: {English MT}, type: {type(English MT)}")
        return ''
```

```python
import pandas as pd
import numpy as np
from collections import Counter
from scipy.sparse import csr_matrix
from sklearn.feature_extraction.English MT import TfidfVectorizer
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')  # تأكد من تحميل قائمة التوقف
import matplotlib.pyplot as plt
from wordcloud import WordCloud

class PropagandaEDA:
    def __init__(self, df, English MT_column, label_column):
        self.df = df
        self.English MT_column = English MT_column
        self.label_column = label_column
        self.stats = {}  # To store statistics and results
        self.resampled_English MTs = None  # This will hold the English MTs after resampling

    def preprocess_data(self):
        """Preprocess the data (e.g., handle missing values, etc.)."""
        # Drop missing values in the English MT column
        self.df.dropna(subset=[self.English MT_column], inplace=True)
        return self

    def analyze_labels(self):
        """Analyze label distribution."""
        label_counts = self.df[self.label_column].value_counts(normalize=True) * 100
        self.stats['label_distribution'] = label_counts
        return self

    def analyze_resampled_labels(self, X_res, y_res, vectorizer=None):
        """Analyze resampled labels after balancing."""
        if vectorizer is None:
            raise ValueError("The 'vectorizer' parameter is required to decode feature indices.")

        # Convert sparse matrix to list of strings if necessary
        if isinstance(X_res, csr_matrix):
            feature_names = vectorizer.get_feature_names_out()
            self.resampled_English MTs = [
                " ".join(feature_names[i] for i in row.indices)
                for row in X_res
            ]
        elif isinstance(X_res, (np.ndarray, list)):
            self.resampled_English MTs = list(X_res)
        else:
            raise ValueError("X_res should be a list, numpy array, or sparse matrix of English MT data.")

        self.stats['resampled_label_distribution'] = Counter(y_res)
        return self

    def analyze_vocabulary(self):
        """Analyze vocabulary of the resampled English MTs with stopwords removed automatically by TfidfVectorizer."""
        if self.resampled_English MTs is None:
            raise ValueError("Resampled English MTs not available. Please run analyze_resampled_labels() first.")

        # Combine stopwords from English, Spanish, and French
        stop_words = list(set(stopwords.words('english')) | set(stopwords.words('spanish')) | set(stopwords.words('french')))

        # Use TfidfVectorizer to process English MTs and remove stopwords
        vectorizer = TfidfVectorizer(stop_words=stop_words)
        tfidf_matrix = vectorizer.fit_transform(self.resampled_English MTs)

        # Get the vocabulary and its frequencies
        feature_names = vectorizer.get_feature_names_out()
        word_freq = tfidf_matrix.sum(axis=0).A1  # Convert to dense format to sum over columns
        word_freq = dict(zip(feature_names, word_freq))

        # Sort by frequency
        sorted_word_freq = sorted(word_freq.items(), key=lambda item: item[1], reverse=True)

        # Store vocabulary stats
        self.stats['total_words_resampled'] = sum(word_freq.values())
        self.stats['unique_words_resampled'] = len(word_freq)
        self.stats['vocabulary_richness_resampled'] = len(word_freq) / sum(word_freq.values())
        self.stats['top_words_resampled'] = sorted_word_freq[:10]

        # Step 4: Visualize with WordCloud
        plt.figure(figsize=(12, 8))
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('Word Cloud of Resampled English MTs (Stop Words Removed)')
        plt.show()

        # Return top words for analysis
        self.stats['top_filtered_words_resampled'] = sorted_word_freq[:10]
        return self  # Ensure method chaining

    def generate_report(self):
        """Generate and display the report."""
        print("Label Distribution:")
        print(self.stats.get('label_distribution'))
        print("\nResampled Label Distribution:")
        print(self.stats.get('resampled_label_distribution'))
        print("\nVocabulary Stats:")
        print(f"Total Words: {self.stats.get('total_words_resampled')}")
        print(f"Unique Words: {self.stats.get('unique_words_resampled')}")
        print(f"Vocabulary Richness: {self.stats.get('vocabulary_richness_resampled')}")
        print(f"Top 10 Words: {self.stats.get('top_words_resampled')}")
        print(f"Top 10 Filtered Words: {self.stats.get('top_filtered_words_resampled')}")
        return self


# ==========================================
# افتراضات:
# 1- df معرف مسبقاً ويحتوي على العمودين "English MT" و "Type of Propaganda"
# 2- تم تنفيذ preprocessing على df
# 3- تم تقسيم البيانات إلى تدريب واختبار
# 4- تم عمل vectorization باستخدام vectorizer على بيانات التدريب
# 5- تم تنفيذ ال Resampling (على سبيل المثال RandomOverSampler) للحصول على X_res, y_res
# 6- يتم استدعاء الكود الآتي بعد تنفيذ جميع الخطوات السابقة
# ==========================================

# على سبيل المثال:
# X_res, y_res نتائج عملية Resampling
# vectorizer نفس الـVectorizer اللي دربته على بيانات التدريب

eda = PropagandaEDA(df, English MT_column="English MT", label_column="Type of Propaganda")

eda.preprocess_data() \
   .analyze_labels() \
   .analyze_resampled_labels(X_res, y_res, vectorizer=vectorizer) \
   .analyze_vocabulary() \
   .generate_report()
```

```python
import pandas as pd
import numpy as np
from collections import Counter
from scipy.sparse import csr_matrix
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
from sklearn.feature_extraction.English MT import TfidfVectorizer

# -------------------------------------------
# تعريف كلاس EDA (كما هو بعد التعديلات السابقة)
class PropagandaEDA:
    def __init__(self, df, English MT_column, label_column):
        self.df = df
        self.English MT_column = English MT_column
        self.label_column = label_column
        self.stats = {}
        self.resampled_English MTs = None

    def preprocess_data(self):
        self.df.dropna(subset=[self.English MT_column], inplace=True)
        return self

    def analyze_labels(self):
        label_counts = self.df[self.label_column].value_counts(normalize=True) * 100
        self.stats['label_distribution'] = label_counts
        return self

    def analyze_resampled_labels(self, X_res, y_res, vectorizer=None):
        if vectorizer is None:
            raise ValueError("The 'vectorizer' parameter is required to decode feature indices.")

        if isinstance(X_res, csr_matrix):
            feature_names = vectorizer.get_feature_names_out()
            self.resampled_English MTs = [
                " ".join(feature_names[i] for i in row.indices)
                for row in X_res
            ]
        elif isinstance(X_res, (np.ndarray, list)):
            self.resampled_English MTs = list(X_res)
        else:
            raise ValueError("X_res should be a list, numpy array, or sparse matrix.")

        self.stats['resampled_label_distribution'] = Counter(y_res)
        return self

    def analyze_vocabulary(self):
        if self.resampled_English MTs is None:
            raise ValueError("Resampled English MTs not available. Run analyze_resampled_labels first.")

        stop_words_list = list(set(stopwords.words('english')) | set(stopwords.words('spanish')) | set(stopwords.words('french')))
        vectorizer = TfidfVectorizer(stop_words=stop_words_list)
        tfidf_matrix = vectorizer.fit_transform(self.resampled_English MTs)

        feature_names = vectorizer.get_feature_names_out()
        word_freq = tfidf_matrix.sum(axis=0).A1
        word_freq = dict(zip(feature_names, word_freq))

        sorted_word_freq = sorted(word_freq.items(), key=lambda item: item[1], reverse=True)
        self.stats['total_words_resampled'] = sum(word_freq.values())
        self.stats['unique_words_resampled'] = len(word_freq)
        self.stats['vocabulary_richness_resampled'] = len(word_freq) / sum(word_freq.values())
        self.stats['top_words_resampled'] = sorted_word_freq[:10]

        plt.figure(figsize=(12, 8))
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('Word Cloud of Resampled English MTs (Stop Words Removed)')
        plt.show()

        self.stats['top_filtered_words_resampled'] = sorted_word_freq[:10]
        return self

    def generate_report(self):
        print("Label Distribution:")
        print(self.stats.get('label_distribution'))
        print("\nResampled Label Distribution:")
        print(self.stats.get('resampled_label_distribution'))
        print("\nVocabulary Stats:")
        print(f"Total Words: {self.stats.get('total_words_resampled')}")
        print(f"Unique Words: {self.stats.get('unique_words_resampled')}")
        print(f"Vocabulary Richness: {self.stats.get('vocabulary_richness_resampled')}")
        print(f"Top 10 Words: {self.stats.get('top_words_resampled')}")
        print(f"Top 10 Filtered Words: {self.stats.get('top_filtered_words_resampled')}")
        return self

# -------------------------------------------
# تعريف كلاس FeatureEngineering (كما هو)
class FeatureEngineering:
    def __init__(self, df, eda_obj, English MT_column):
        self.df = df
        self.English MT_column = English MT_column
        self.eda = eda_obj
        self.vectorizer = None

    def filter_keywords(self, word_freq, keywords):
        filtered_freq = {word: freq for word, freq in word_freq.items() if word in keywords}
        return filtered_freq

    def build_vocabulary(self):
        top_filtered_words = self.eda.stats.get('top_filtered_words_resampled', [])
        if not top_filtered_words:
            raise ValueError("No valid keywords found in the EDA results.")
        filtered_keywords = [word for word, _ in top_filtered_words]
        return filtered_keywords

    def visualize_vocabulary(self, word_freq):
        if not word_freq:
            print("No valid keywords found.")
            return
        plt.figure(figsize=(12, 8))
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('Word Cloud of Filtered Vocabulary')
        plt.show()

    def perform_feature_engineering(self):
        keywords = self.build_vocabulary()
        # استخدام نفس الإعدادات، لكن هنا سيختلف عن EDA لأننا لا نحتاج سوى لاستخراج الميزات
        self.vectorizer = TfidfVectorizer(stop_words='english')
        tfidf_matrix = self.vectorizer.fit_transform(self.df[self.English MT_column])
        feature_names = self.vectorizer.get_feature_names_out()
        word_freq = dict(zip(feature_names, tfidf_matrix.sum(axis=0).A1))
        filtered_word_freq = self.filter_keywords(word_freq, keywords)
        self.visualize_vocabulary(filtered_word_freq)
        return tfidf_matrix, filtered_word_freq

# -------------------------------------------
# مثال لتطبيق الخطوات بالترتيب الصحيح
# تأكد أن df معرف ويحتوي على العمودين: "English MT", "Type of Propaganda"
# 1- تقسيم البيانات
X_train, X_test, y_train, y_test = train_test_split(df['English MT'], df['Type of Propaganda'],
                                                    test_size=0.2,
                                                    random_state=42,
                                                    stratify=df['Type of Propaganda'])

# 2- Vectorizer على بيانات التدريب فقط
vectorizer = TfidfVectorizer(max_features=5000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# 3- Resampling على بيانات التدريب فقط
ros = RandomOverSampler(random_state=42)
X_res, y_res = ros.fit_resample(X_train_vec, y_train)

# 4- استخدام الـ EDA على البيانات المعاد توازنها
eda = PropagandaEDA(df, English MT_column="English MT", label_column="Type of Propaganda")
eda.preprocess_data() \
   .analyze_labels() \
   .analyze_resampled_labels(X_res, y_res, vectorizer=vectorizer) \
   .analyze_vocabulary() \
   .generate_report()

# 5- الآن FeatureEngineering باستخدام نفس الـ EDA
fe = FeatureEngineering(df, eda_obj=eda, English MT_column="English MT")
X_features, filtered_word_freq = fe.perform_feature_engineering()
print("Filtered Word Frequencies:", filtered_word_freq)
```

```python
# Initialize BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
model = BertModel.from_pretrained('bert-base-multilingual-cased')

# Move model to GPU if available
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
model.to(device)
model.eval()

def get_bert_embeddings(English MTs, tokenizer, model, batch_size=32):
    embeddings = []
    for i in tqdm(range(0, len(English MTs), batch_size)):
        batch_English MTs = English MTs[i:i+batch_size]
        encoded_input = tokenizer(batch_English MTs, padding=True, truncation=True, return_tensors='pt', max_length=512)
        encoded_input = {key: value.to(device) for key, value in encoded_input.items()}

        with torch.no_grad():
            model_output = model(**encoded_input)

        # Mean pooling
        attention_mask = encoded_input['attention_mask']
        token_embeddings = model_output.last_hidden_state
        input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
        sum_embeddings = torch.sum(token_embeddings * input_mask_expanded, 1)
        sum_mask = torch.clamp(input_mask_expanded.sum(1), min=1e-9)
        batch_embeddings = sum_embeddings / sum_mask
        embeddings.append(batch_embeddings.cpu().numpy())

    return np.vstack(embeddings)

bert_embeddings = get_bert_embeddings(English MTs, tokenizer, model)
```

```python
# Combine TF-IDF features with BERT embeddings
X_combined = np.hstack([X_res.toarray(), bert_embeddings])
```

```python
le = LabelEncoder()
y_res_encoded = le.fit_transform(y_res)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, y_res_encoded, test_size=0.2, stratify=y_res_encoded, random_state=42
)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
print("\nTraining Random Forest with BERT features...")
rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)
y_proba = rf_model.predict_proba(X_test) if hasattr(rf_model, "predict_proba") else None

accuracy = accuracy_score(y_test, y_pred)
print("Random Forest with BERT Features Results:")
print("Accuracy:", accuracy)

if y_proba is not None:
    try:
        roc_auc = roc_auc_score(y_test, y_proba, multi_class='ovr')
        print("ROC AUC Score:", roc_auc)
    except ValueError as e:
        print(f"Error calculating ROC AUC: {e}")
else:
    print("ROC AUC Score: N/A")

print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from transformers import XLMRobertaTokenizer, XLMRobertaForSequenceClassification, Trainer, TrainingArguments
import torch
from torch.utils.data import Dataset
from tqdm import tqdm
```

```python
class English MTDataset(Dataset):
    def __init__(self, English MTs, labels, tokenizer, max_len=128):
        self.English MTs = English MTs
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.English MTs)

    def __getitem__(self, idx):
        English MT = str(self.English MTs[idx])
        label = self.labels[idx]
        encoding = self.tokenizer.encode_plus(
            English MT,
            add_special_tokens=True,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }
```

```python
tokenizer = XLMRobertaTokenizer.from_pretrained('xlm-roberta-base')

train_dataset = English MTDataset(
    English MTs=X_train,
    labels=y_train,
    tokenizer=tokenizer,
    max_len=128
)

test_dataset = English MTDataset(
    English MTs=X_test,
    labels=y_test,
    tokenizer=tokenizer,
    max_len=128
)
```

```python
model = XLMRobertaForSequenceClassification.from_pretrained(
    'xlm-roberta-base',
    num_labels=len(le.classes_)
)
```

```python
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
    greater_is_better=True
)
```

```python
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    acc = accuracy_score(labels, preds)
    return {'accuracy': acc}
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics
)
```

```python
predictions = trainer.predict(test_dataset)
preds = np.argmax(predictions.predictions, axis=1)
accuracy = accuracy_score(y_test, preds)
print("XLM-RoBERTa Fine-Tuned Results:")
print("Accuracy:", accuracy)
print(classification_report(y_test, preds, target_names=le.classes_))

# Calculate ROC AUC if possible
if predictions.predictions.ndim > 1:
    try:
        roc_auc = roc_auc_score(y_test, predictions.predictions, multi_class='ovr')
        print("ROC AUC Score:", roc_auc)
    except ValueError as e:
        print(f"Error calculating ROC AUC: {e}")
else:
    print("ROC AUC Score: N/A")
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,
    n_jobs=-1,
    verbose=2,
    scoring='accuracy'
)

grid_search.fit(X_train, y_train)
print("Best Parameters:", grid_search.best_params_)

best_rf = grid_search.best_estimator_
y_pred_best = best_rf.predict(X_test)
accuracy_best = accuracy_score(y_test, y_pred_best)
print("Best Random Forest Accuracy:", accuracy_best)
print(classification_report(y_test, y_pred_best, target_names=le.classes_))
```
